```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')
```

```python
salary_df = pd.read_csv('Downloads\ds_salaries.csv')
```

```python
entry_df = salary_df[salary_df.experience_level == 'EN']
medium_df = salary_df[salary_df.experience_level == 'MI']
senior_df = salary_df[salary_df.experience_level == 'SE']
exec_df = salary_df[salary_df.experience_level == 'EX']
```

```python
plt.figure(figsize=(20,10))

plt.hist(entry_df.salary_in_usd, alpha=0.4, bins=np.arange(1000, 200000, 2500));
plt.hist(medium_df.salary_in_usd, alpha=0.4, bins=np.arange(1000, 200000, 2500));
plt.hist(senior_df.salary_in_usd, alpha=0.4, bins=np.arange(1000, 200000, 2500));
plt.hist(exec_df.salary_in_usd, alpha=0.4, bins=np.arange(1000, 200000, 2500));

plt.legend(['Entry', 'Mid', 'Senior', 'Executive'])

plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.title=('Salaries by Work Type & Experience')
sns.scatterplot(x=salary_df.salary_in_usd,
                y=salary_df.remote_ratio,
                hue=salary_df.experience_level,
                s=100)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='experience_level', y='salary_in_usd',hue='remote_ratio', data=salary_df);
plt.show()
```

```python
title_df = salary_df[['job_title', 'salary_in_usd']]
title_df
```

```python
title_df = title_df.groupby('job_title').mean()
```

```python
title_df = title_df.sort_values('salary_in_usd', ascending=True)
```

```python
ax = title_df.plot(kind='barh', figsize=(10, 40),width=0.7, title='Salaries per Job Title')
ax.legend().remove()
plt.show()
```
